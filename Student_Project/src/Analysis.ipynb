```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../Data_Clean/cleaned_data.csv")
# Strip whitespaces from column names
df.columns = df.columns.str.strip()

gender_colors = {'male': 'rosybrown', 'female': 'navy'}


# Visualization 1: Scatter plot - Math scores vs. reading scores, colored by gender
plt.figure(figsize=(8, 6))
sns.scatterplot(x='math score', y='reading score', hue='gender', palette=gender_colors,data=df)
plt.title('Math Scores vs. Reading Scores (Colored by Gender)')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../Data_Clean/cleaned_data.csv")
# Strip whitespaces from column names
df.columns = df.columns.str.strip()
race_colors = {'group A': 'midnightblue', 'group B': 'lime', 'group C': 'crimson', 'group D':'maroon', 'group E':'lightseagreen'}

# 2. Bar Chart
sns.barplot(data=df, x='gender', y='math score', hue='race/ethnicity', palette=race_colors)
plt.title('Average Math Score by Gender and Race/Ethnicity')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../Data_Clean/cleaned_data.csv")
# Strip whitespaces from column names
df.columns = df.columns.str.strip()
education_colors = {'high school': 'blue', 'some college': 'plum', "bachelor's degree":'lightslategray', "master's degree":'teal',
                     "associate's degree":'peru','some high school':'crimson'}

# 3. Box Plot
sns.boxplot(data=df, x='parental level of education', y='math score', palette=education_colors)
plt.xticks(rotation=45)
plt.title('Distribution of Math Scores by Parental Education Level')
plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../Data_Clean/cleaned_data.csv")
# Strip whitespaces from column names
df.columns = df.columns.str.strip()
# 4. Stacked Bar Chart
table = pd.crosstab(df['parental level of education'], df['lunch'])
table.div(table.sum(1), axis=0).plot(kind='bar', stacked=True, color=['gray', 'black'])
plt.title('Proportion of Lunch Status by Parental Education Level')
plt.xlabel('Parental Education Level')
plt.ylabel('Proportion')
plt.xticks(rotation=45)
plt.show()
```
